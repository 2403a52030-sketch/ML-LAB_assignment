```python
import pandas as pd
dataset = pd.read_csv("/content/sample_data/Salary_dataset.csv")
def predict(X, m, c):
    return m * X + c
```

```python
import pandas as pd
dataset = pd.read_csv("/content/sample_data/Salary_dataset.csv")
def cost_function(y, y_pred):
    return np.mean((y - y_pred) ** 2)
```

```python
import pandas as pd
dataset = pd.read_csv("/content/sample_data/Salary_dataset.csv")
# Correctly define X (YearsExperience) and y (Salary) for simple linear regression
X = dataset['YearsExperience'].values.reshape(-1, 1)
y = dataset['Salary'].values.reshape(-1, 1)

y_pred = predict(X, m, c)
cost = cost_function(y, y_pred)
print("Initial Cost:", cost)
```

```python
import pandas as pd
dataset = pd.read_csv("/content/sample_data/Salary_dataset.csv")
n = len(X)

for _ in range(epochs):
    y_pred = predict(X, m, c)
    dm = (-2/n) * np.sum(X * (y - y_pred))
    dc = (-2/n) * np.sum(y - y_pred)
    m = m - learning_rate * dm
    c = c - learning_rate * dc

print("Final Slope:", m)
print("Final Intercept:", c)
```

```python
import pandas as pd
dataset = pd.read_csv("/content/sample_data/Salary_dataset.csv")
final_predictions = predict(X, m, c)
mse = cost_function(y, final_predictions)
print("Mean Squared Error:", mse)
```
